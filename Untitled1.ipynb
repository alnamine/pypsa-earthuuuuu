```python
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import warnings
import requests
import os

# --- 0. Suppress Warnings ---
warnings.filterwarnings("ignore", category=UserWarning)
print("Libraries imported.")

# --- 1. Define File Paths and Bounding Box ---
wind_file = "resources/renewable_profiles/profile_onwind.nc"
local_shapefile = "algeria_wilayas.geojson" # Uses the file we already downloaded

# --- ADDED Béjaïa to the list ---
wilaya_names = ['Jijel', 'Constantine', 'Mila', 'Skikda', 'Béjaïa']

# --- WIDENED the map bounds to include Béjaïa ---
map_bounds = [4.9, 7.1, 36.1, 37.2] # [lon_min, lon_max, lat_min, lat_max]

# --- 2. Load Data ---
print("Loading wind potential and border data...")
ds_wind = xr.open_dataset(wind_file)
all_wilayas = gpd.read_file(local_shapefile)

# --- 3. Process Data ---
print(f"Filtering for: {wilaya_names}")
wilayas_to_plot = all_wilayas[
    all_wilayas['shapeName'].isin(wilaya_names)
]

if wilayas_to_plot.empty:
    print(f"ERROR: Could not find wilayas {wilaya_names} in the shapefile.")
else:
    print(f"Found {len(wilayas_to_plot)} wilayas to plot.")

    # --- 4. Create the Map ---
    print("Setting up the map...")
    projection = ccrs.PlateCarree()
    fig, ax = plt.subplots(figsize=(12, 10), subplot_kw={"projection": projection})

    # --- Plot Wind Potential (in Blue) ---
    print("Plotting wind potential...")
    ds_wind.potential.plot(
        ax=ax, 
        transform=projection, 
        cmap='Blues',
        cbar_kwargs={'label': 'Onshore Wind Potential (MW per grid cell)'}
    )

    # --- Plot Wilaya Borders ---
    print("Plotting wilaya borders...")
    wilayas_to_plot.plot(
        ax=ax,
        transform=projection,
        edgecolor='black',
        facecolor='none',
        linewidth=1.5
    )

    # --- Add Wilaya Names ---
    print("Adding labels...")
    for idx, row in wilayas_to_plot.iterrows():
        center = row.geometry.centroid
        ax.text(
            center.x, 
            center.y, 
            row['shapeName'],
            transform=projection,
            color='black',
            fontsize=12,
            fontweight='bold',
            ha='center',
            va='center',
            bbox=dict(facecolor='white', alpha=0.4, pad=0.1, boxstyle='round,pad=0.2')
        )

    # --- Set Map Extent and Final Touches ---
    ax.set_extent(map_bounds, crs=projection)
    ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
    plt.title("High-Res Onshore Wind Potential (2024 Data)")
    plt.savefig("jijel_region_WIND_MAP_ALL.png", dpi=300)

    print("\n--- DONE! ---")
    print("Map saved as 'jijel_region_WIND_MAP_ALL.png'")
```
