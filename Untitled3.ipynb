```python
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.img_tiles as cimgt  # <--- NEW IMPORT
import warnings
import requests
import os

# --- 0. Suppress Warnings ---
warnings.filterwarnings("ignore", category=UserWarning)
print("Libraries imported.")

# --- 1. Define File Paths and Bounding Box ---
solar_file = "resources/renewable_profiles/profile_solar.nc"
local_shapefile = "algeria_wilayas.geojson" # Uses the file we already downloaded
wilaya_names = ['Jijel', 'Constantine', 'Mila', 'Skikda', 'Bejaia','SÃ©tif']
map_bounds = [4.9, 7.1, 36.1, 37.2] # [lon_min, lon_max, lat_min, lat_max]

# --- 2. Load Data ---
print("Loading solar potential and border data...")
ds_solar = xr.open_dataset(solar_file)
all_wilayas = gpd.read_file(local_shapefile)
    
# --- 3. Process Data ---
print(f"Filtering for: {wilaya_names}")
wilayas_to_plot = all_wilayas[
    all_wilayas['shapeName'].isin(wilaya_names)
]
print(f"Found {len(wilayas_to_plot)} wilayas to plot.")

# --- 4. Create the Map ---
print("Setting up the map...")
# --- NEW: Create the Stamen terrain tile object ---
terrain_tiles = cimgt.Stamen('terrain-background')

projection = ccrs.PlateCarree()
fig, ax = plt.subplots(figsize=(12, 10), subplot_kw={"projection": projection})

# --- NEW: Add the terrain tiles to the map FIRST ---
# Zoom level 8 or 9 is good for this region
print("Adding topography background...")
ax.add_image(terrain_tiles, 9)

# --- Plot Solar Potential (with transparency) ---
print("Plotting solar potential...")
ds_solar.potential.plot(
    ax=ax, 
    transform=projection, 
    cmap='inferno',
    alpha=0.7,  # <--- NEW: Add 70% transparency to see terrain
    cbar_kwargs={'label': 'Solar Potential (MW per grid cell)'}
)

# --- Plot Wilaya Borders ---
print("Plotting wilaya borders...")
wilayas_to_plot.plot(
    ax=ax,
    transform=projection,
    edgecolor='white',
    facecolor='none',
    linewidth=1.5
)

# --- Add Wilaya Names ---
print("Adding labels...")
for idx, row in wilayas_to_plot.iterrows():
    center = row.geometry.centroid
    ax.text(
        center.x, 
        center.y, 
        row['shapeName'],
        transform=projection,
        color='white',
        fontsize=12,
        fontweight='bold',
        ha='center',
        va='center',
        bbox=dict(facecolor='black', alpha=0.4, pad=0.1, boxstyle='round,pad=0.2')
    )

# --- Set Map Extent and Final Touches ---
ax.set_extent(map_bounds, crs=projection)
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
ax.coastlines(color='white', linewidth=0.5) # Add coastlines for clarity
plt.title("High-Res Solar Potential with Topography (2024 Data)")
plt.savefig("jijel_region_SOLAR_MAP_TOPO.png", dpi=300)

print("\n--- DONE! ---")
print("Map saved as 'jijel_region_SOLAR_MAP_TOPO.png'")
```
